```python
df2=pd.read_csv("testData.csv")
df2.head(10)
```

```python
df3=df1.append(df2)
df3.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder() 
df3["Gender"]=le.fit_transform(df3['Gender'].astype(str))
df3["Married"]=le.fit_transform(df3['Married'].astype(str))
df3["Dependents"]=le.fit_transform(df3['Dependents'].astype(str))
df3["Self_Employed"]=le.fit_transform(df3['Self_Employed'].astype(str))
df3["Property_Area"]=le.fit_transform(df3['Property_Area'].astype(str))
df3.head(10)
```

```python
import random as r
import numpy as np
dfc = df3.copy()

dfc['LoanAmount'].fillna((dfc['LoanAmount'].median()), inplace=True)
dfc['Dependents'].fillna((dfc['Dependents'].median()), inplace=True)
dfc['Loan_Amount_Term'].fillna((dfc['Loan_Amount_Term'].mean()), inplace=True)
dfc.isna().sum()
```

```python
X=df3[['ApplicantIncome']]
y=df3[['Married']]
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnModel = KNeighborsClassifier(n_neighbors=5)
knnModel.fit(X,y) 
```

```python
knnModel.score(X,y)*100
```

```python
from sklearn.metrics import accuracy_score
ypredict = knnModel.predict(X)
print('Accuracy score is',accuracy_score(y,ypredict)*100)
```
